```python
df = df.drop(labels={'EmpID','EmpName','BiasInfluentialFactor','GraduationYear', }, axis=1)
df
```

```python
categoricial_features = ['LanguageOfCommunication','Gender','JobProfileIDApplyingFor','HighestDegree','DegreeBranch','GraduatingInstitute','MartialStatus','CurrentDesignation','CurrentCompanyType','DepartmentInCompany']
df = pd.get_dummies(df, columns=categoricial_features)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
correlation = df.corr() 
fig = plt.figure(figsize = (20, 20)) 
sns.heatmap(correlation, vmax = 1, square = True, cmap="YlGnBu") 
plt.show()
```

```python
df.describe()
```

```python
numeric_features = ['Age','LatestDegreeCGPA','YearsOfExperince','CurrentCTC','ExpectedCTC','EmpScore','TotalLeavesTaken']
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(df[numeric_features])
df[numeric_features] = ss.transform(df[numeric_features])
df
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split
df1 = df.drop(labels='FitmentPercent', axis=1)

X= df1  
y=df['FitmentPercent']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
from sklearn import datasets, linear_model, metrics
clf = linear_model.LinearRegression()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

clf.score(X_test, y_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import BayesianRidge

model = BayesianRidge()
model.fit(X_train, y_train)
  
# Model making a prediction on test data
prediction = model.predict(X_test)
  
# Evaluation of r2 score of the model against the test set
print(f"r2 Score Of Test Set : {r2_score(y_test, prediction)}")
```
